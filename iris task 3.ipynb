```python
colors=['Black','teal','red']
```

```python
species=['Iris-virginica','Iris-setosa','Iris-versicolor']
```

```python
for i in range(3):
    x=iris[iris['species']==species[i]]
    plt.scatter(x['sepal_length'],x['sepal_width'],c=colors[i],label=species[i])
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.legend()
```

```python
for i in range(3):
    x=iris[iris['species']==species[i]]
    plt.scatter(x['petal_length'],x['petal_width'],c=colors[i],label=species[i])
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.legend()
```

```python
numeric_columns=iris.drop(columns='species')
corr=numeric_columns.corr()
fig,axis=plt.subplots(figsize=(5,5))
sns.heatmap(corr,annot=True,ax=axis,cmap='coolwarm')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
la=LabelEncoder()
```

```python
iris['species']=la.fit_transform(iris['species'])
iris.head(16)
```

```python
x=iris.drop(columns='species')
```

```python
y=iris['species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
kn=KNeighborsClassifier()
kn.fit(x_train,y_train)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
LR_accuracy=lr.score(x_test,y_test)*100
KN_accuracy=kn.score(x_test,y_test)*100
DT_accuracy=dt.score(x_test,y_test)*100
```

```python
print(f"Accuracy value by LogisticRegression:{LR_accuracy}%")
print(f"Accuracy value by K Nearest neighbors:{KN_accuracy}%")
print(f"Accuracy value by Decision Tree classifer:{DT_accuracy}%")
```
